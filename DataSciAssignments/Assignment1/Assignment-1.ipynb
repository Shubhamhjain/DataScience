```python
#Question 9

def remove_duplicates_and_reverse(arr):
    arr1=[]              #initializing an empty array
    for i in arr:
        if i in arr1:    #checking if the element exists in arr1; if it exists then we move to the next element else we add that element to the arr1
            continue
        else:
            arr1.append(i)
    return arr1[::-1]    #returning the reverse order of arr1

arr=[1,1,2,2,3,3,4,4,5,6,7,7]
remove_duplicates_and_reverse(arr)
            
```

```python
#Question11

def check_if_palindrome(inputStr):
    return inputStr==inputStr[::-1]         # reversing the array and matching the two strings

def check_if_symmetrical(inputStr):
    if(len(inputStr)%2==0):
        firstHalf= inputStr[:int(len(inputStr)/2)]           # splitting strings in 2 halves and comparing the halves
        secondHalf= inputStr[int(len(inputStr)/2):]
        if(firstHalf==secondHalf):
            return True
    return False

def check_result(inputStr):                   #function for checking whether string is palindrome,symmetrical or not
    if check_if_palindrome(inputStr):
        print('String is a Palindrome.')
    elif check_if_symmetrical(inputStr):
        print('String is a Symmetrical.')
    else:
        print('String is not a Palindrome nor Symmetrical.')
        

str1 = 'malayalam'
str2='byebye'

print("String Entered: " + str1 + "\n")
check_result(str1)

print('\n')

print("String Entered: " + str2 + "\n")
check_result(str2)
```

```python
# Question12

def Fibonacci(n):
    if n<0:                       #if the entered number is less than 0 then it is incorrect
        print('incorrect length')
    elif n<2:                     # if n is 1 then F(1) is 1, if n is 0 then F(0) is 0
        return n
    else:
        return Fibonacci(n-1) + Fibonacci(n-2)  # recursive call to compute the fibonacci 

Fibonacci(5)
```

```python
# Question13

import time
start_time = time.time()     #starting the timer

def LongestCommonSubstring(str1,str2):
    matrix = [[0]*(1+len(str2)) for s in range(len(str1)+1)]    # initializing a matrix with 0 of length of both the strings
    longest,longest1=0,0                                        #declaring variables for computing the longest substring
    for i in range(1,len(str1)+1):                              #outer loop for string 1
        for j in range(1,len(str2)+1):                          #inner loop for string 2
            if(str1[i-1]==str2[j-1]):                           #condition where the characters of both strings gets matched
                matrix[i][j]= matrix[i-1][j-1]+1                #1 is added to the previous value of the matrix and stored in the current index 
                if(matrix[i][j]>longest):
                    longest=matrix[i][j]                        #variable for storing the maximum value
                    longest1=i
    return str1[longest1-longest:longest1]                      #computing the longest common substring
             
str1 = "There is a lady on the mountain with an umbrella" 
str2 = "I love to look at the mountain when it is snowcapped"
print(LongestCommonSubstring(str1,str2))
print("--- %s seconds ---" % (time.time() - start_time))        #displaying the time taken for the code to run
```

```python
#using mean, median and mode we can fill the missing values
#for float values ad integers(numerical values), we use mean and median
#for categorical data , we use mode


#below are the columns containing numerical values

ds['Employment_Info_1'] = ds['Employment_Info_1'].fillna(ds['Employment_Info_1'].mean())
ds['Employment_Info_4'] = ds['Employment_Info_4'].fillna(ds['Employment_Info_4'].mean())
ds['Employment_Info_6'] = ds['Employment_Info_6'].fillna(ds['Employment_Info_6'].mean())
ds['Employment_Info_5'] = ds['Employment_Info_5'].fillna(ds['Employment_Info_5'].mean())
ds['Insurance_History_5'] = ds['Insurance_History_5'].fillna(ds['Insurance_History_5'].mean())
ds['Family_Hist_2'] = ds['Family_Hist_2'].fillna(ds['Family_Hist_2'].mean())
ds['Family_Hist_3'] = ds['Family_Hist_3'].fillna(ds['Family_Hist_3'].mean())
ds['Family_Hist_4'] = ds['Family_Hist_4'].fillna(ds['Family_Hist_4'].mean())
ds['Family_Hist_5'] = ds['Family_Hist_5'].fillna(ds['Family_Hist_5'].mean())

```

```python
#below are the columns containing categorical values

ds['Medical_History_1'] = ds['Medical_History_1'].fillna(ds['Medical_History_1'].mode())
ds['Medical_History_10'] = ds['Medical_History_10'].fillna(ds['Medical_History_10'].mode())
ds['Medical_History_15'] = ds['Medical_History_15'].fillna(ds['Medical_History_15'].mode())
ds['Medical_History_24'] = ds['Medical_History_24'].fillna(ds['Medical_History_24'].mode())
ds['Medical_History_32'] = ds['Medical_History_32'].fillna(ds['Medical_History_32'].mode())


```

```python
#multicollinearity can help us discard similar variabls
print(ds.corr())

#Here a correlation matrix is generated consisting of 126 rows and 126 columns with values between -1 to 1.
#Values closer to -1 or 1 mean that the variables are highly correlated.
#Values closer to 0 mean that values are less correlated.


#here calculating vif for each feature. Features with similar vif are highly correlated
vif_dataset = add_constant(ds)
ds=pd.Series([variance_inflation_factor(vif_dataset.values, i)
               for i in range(vif_dataset.shape[1])],
              index=vif_dataset.columns)
```
